```python
movies_stats = df.groupby('title').agg({'rating': ['mean', 'count']})
```

```python
movies_stats.columns = ['avg_rating', 'num_rating']
movies_stats = movies_stats.reset_index()
```

```python
movies_stats.head()
```

```python
popular_movies = movies_stats[movies_stats['num_rating'] > 100]
popular_movies
```

```python
popular_movies = popular_movies.sort_values(by='avg_rating', ascending=False)

def getPopularMovies(n=10):
    return popular_movies.head(n)
```
